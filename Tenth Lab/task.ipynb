<h2>Этап 1. Подготовка данных</h2>
<h4>Скачайте файл с набором данных событий информационной безопасности в
формате JSON (events.json).<br>
Важно: можно сгенерировать данные самостоятельно или использовать
предоставленные.</h4>

```python
import json
import pandas as pd
path_json = "src/events.json"
```

<h2>Этап 2. Анализ данных</h2>
<h4>Загрузите данные из файла JSON в датафрейм с помощью Pandas и
проанализируйте их. <br>Вы должны изучить распределение событий
информационной безопасности по их типам (по полю «signature»).</h4>

```python
with open(path_json, "r") as file:
    data = json.load(file)
    df = pd.DataFrame(data["events"])
#Метод самого DataFrame для подсчёта уникальных значений
signature_counts = df['signature'].value_counts()
print(signature_counts)
#print(df)
```

<h2>Этап 3. Визуализация данных</h2>
<h4>Постройте график распределения типов событий информационной
безопасности. <br>Используйте Seaborn или Matplotlib для создания визуализации.</h4>

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,9))
signature_counts.plot(kind='bar')
plt.title("Распределение сигнатур")
plt.ylabel("Количество")
plt.xlabel("Имя сигнатуры")
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```
