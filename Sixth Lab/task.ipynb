<h3>Нужно реализовать класс Account, который отражает абстракцию базового
поведения банковского аккаунта:<br></h3>
● создать банковский аккаунт с параметрами: имя; стартовый баланс с
которым зарегистрирован аккаунт; история операций*;<br>
● реализовать два метода, которые позволяют положить деньги на счёт
или снять деньги со счёта;<br>
● продумать, как можно хранить историю поступления или снятия
денег, чтобы с ней было удобно работать*.

```python
from decimal import *
class Account:
    def __init__(self, name: str, money: Decimal): # Конструктор класса
      self.name = name
      self.money = money
      self.history = ["Добавлен пользователь"]

    def write_history(self, value: Decimal, description: str):
       self.history.append(f"Сумма: {value}, Баланс: {self.money} {description}" )
         # "value": value,
        #  "balance_after": self.money,
       #   "description": description
     #  })

    def look_history(self):
       print(self.history)

    def get_money(self):
       return self.money
    
    def get_name(self):
       return self.name
    
    # Первый метод для добавления денег
    def add_money(self, value: Decimal, description: str = ""):
       self.money = self.get_money() + value
       if description == "":
          self.write_history(value, f"Внесение {value} на счёт")
       else:
          self.write_history(value, description)
    
    # Первый метод для снятия денег
    def take_money(self, value: Decimal, description: str = ""):
        self.money = self.get_money() - value
        if self.money <= 0:
            self.money = self.money + value
            raise ValueError("Недостаточно средств на балансе для совершения операции")
        else:
            if description == "":
                self.write_history(value, f"Списание {value} со счёта")
            else:
                self.write_history(value, description)
# Второй метод для снятия и добавления денег на баланс
def transfer(from_acc: Account, to_acc: Account, value: Decimal):
    if value <= 0:
       raise ValueError("Перевод не может быть меньше или равен нулю")
    try:
        from_acc.take_money(value, f"Перевод пользователю {to_acc.get_name()} суммы {value}")
    except(ValueError):
       # Решил тут не отображать эту ошибку в истории, т.к. операция не была совершена
       print("Не хватает средств на балансе")
    else:
        to_acc.add_money(value, f"Перевод от {to_acc.get_name()} суммы {value}")
   
User1 = Account("Вася Пупкин", 0.0)
User2 = Account("Игнат Василий", 500123.0)
User1.look_history()
User1.add_money(555.00)
User1.look_history()
User1.take_money(333.0)
User1.look_history()
transfer(User2, User1, 50000.00)
User2.look_history()
transfer(User1, User2, 100000000.00)
User1.look_history()
```
